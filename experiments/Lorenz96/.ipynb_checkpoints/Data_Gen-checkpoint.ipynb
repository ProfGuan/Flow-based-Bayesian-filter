```python
import random
import numpy as np
import os
import math
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from mpl_toolkits.axes_grid1 import make_axes_locatable
from State_Space_Model import Stochastic_Lorenz96
from scipy.stats import multivariate_normal
from tqdm import trange, tqdm

import warnings
warnings.filterwarnings("ignore")
```

```python
######################################
########## dataset parameters ########
######################################
F = 8
delta_t = 0.01
Delta_t = 0.05
N_train = 2000
N_test = 200
T = 500
T_test = 500
r2 = 1
q2 = 0.01
m = 20
n = m

obs_type = 'NL'
dataFolderName = './Data/'
dataFileName = 'Lorenz96_q2={:.3f}_r2={:.3f}_m={}_T={}_obs={}.npy'.format(q2, r2, m, T, obs_type)
data_filename = dataFolderName + dataFileName

if not os.path.exists('./models'):
    os.mkdir('./models')
if not os.path.exists('./Data'):
    os.mkdir('./Data')    
    
######################################
########### Generate Dataset #########
######################################
SSM = Stochastic_Lorenz96(F, q2, r2, delta_t, m, n)
init_state = np.sin(2*np.pi * (np.arange(1, m+1) / m))

if not os.path.exists(data_filename):
    x_train, y_train, x_test, y_test = SSM.generate_sample(delta_t, Delta_t, T, T_test, N_train, N_test, 
                                                           init_state, obs_type)
    np.save(data_filename, np.array([x_train, y_train, x_test, y_test], dtype=object))
else:
    print("Load prepared data...")
    x_train, y_train, x_test, y_test = np.load(data_filename, allow_pickle=True)
```
